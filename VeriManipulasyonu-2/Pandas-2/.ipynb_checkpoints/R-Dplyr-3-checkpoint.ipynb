# R-Dplyr Tarzı Veri Manipülasyonu

* **pip install dfply**

```python
from dfply import *
```

```python
diamonds # kütüphane ile gelen bir fonk
```

```python
df = diamonds.copy()
df.head()
```

```python
df >> head() # dfply kütüphanesi ile gelen özellik
```

```python
# df >>= head() # atama işlemi yaptıK
```

```python
df.head()
```

```python
df >> select(X.carat,X.cut)>>head()
# değişken seçme işlemi
```

```python
df >> select(1,X.price,['x','y']) >> head(3)
# 1 -> değişken seçme yöntemi
# X -> dataframe yakalama durumu
# ['x','y'] -> değişken seçtik
```

```python
df >> select(starts_with('c')) >> head()
# 'c' ile başlayan değişkenleri getirir
```

```python
df >> select(contains('y')) >> head(3)
# 'y' içeren değişkenleri getirir
```

```python
df >> select(columns_between('cut','x')) >> head(3)
# belirli bir değişkenden bir değişkene gider
```

```python
df >> drop(1,X.price,['x','y'])>>head()
# bir değişken siler
```

```python
df >> select(~X.carat,~X.color) >> head(3)
# carat ve color değişkenleri hariç
```

```python
df >> row_slice([10,15])
# 10. ve 15. değerleri aldı
```

```python
df >> row_slice([10])
# 10. değeri aldı
```

```python
df >> group_by('cut') >> row_slice(5)
# cut'a göre grrupladı ve her birinden sonraki 5. değere gitti
```
